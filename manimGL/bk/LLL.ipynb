```python
import numpy as np
import random as rand 
```

```python
rand.random()

rand.randint(1, 19)



x=[rand.gauss(1,2) for i in range(1000)]
y=[rand.gauss(1,2) for i in range(1000)]



fit=np.polyfit(x, y, 3)
fitted=np.polyval(fit, x)

import matplotlib.pyplot as plt

plt.plot(x, y, "*")
plt.plot(x, fitted, ".", "r-")
```
